```python
df = df.drop("Unnamed: 0", axis=1)
df.head(10)
```

```python
df['Inches'].unique()
```

```python
df['Inches']=df['Inches'].str.replace('?', '',regex = False)
```

```python
df['Inches']=pd.to_numeric(df['Inches'], errors='coerce')
```

```python

df['Ram'] = df['Ram'].astype(str).str.replace('GB', '', regex=False)
```

```python
df['Ram']=pd.to_numeric(df['Ram'], errors='coerce')
```

```python
df['Memory']=df['Memory'].str.replace('?','', regex=False)
```

```python
df['Weight']=df['Weight'].str.replace('kg', '')
```

```python
df['Weight']=pd.to_numeric(df['Weight'], errors='coerce') 
df
```

```python
df.isnull().sum()
```

```python
df.isnull().sum().sum()
```

```python
df['Inches']=df['Inches'].fillna(df['Inches'].median())
df['Ram']=df['Ram'].fillna(df['Ram'].median())
df['Weight']=df['Weight'].fillna(df['Weight'].median())
df['Price']=df['Price'].fillna(df['Price'].median())
df
```

```python
df['Cpu']=df['Cpu'].fillna(df['Cpu'].mode()[0])
df['ScreenResolution']=df['ScreenResolution'].fillna(df['ScreenResolution'].mode()[0])
df['Memory']=df['Memory'].fillna(df['Memory'].mode()[0])
df['Company']=df['Company'].fillna(df['Company'].mode()[0])
df['TypeName']=df['TypeName'].fillna(df['TypeName'].mode()[0])
df['Gpu']=df['Gpu'].fillna(df['Gpu'].mode()[0])
df['OpSys']=df['OpSys'].fillna(df['OpSys'].mode()[0])
```

```python
df.isnull().sum()
```

```python
df.shape
```

```python
df.duplicated().sum()
```

```python
df=df.drop_duplicates()
```

```python
df.duplicated().sum()
```

```python
df
```

```python
df["Touchscreen"] = df["ScreenResolution"].apply(lambda x: 1 if "Touchscreen" in x else 0)
df["Ips"] = df["ScreenResolution"].apply(lambda x: 1 if "IPS" in x else 0)
df[["X_resolution", "Y_resolution"]] = df["ScreenResolution"].str.extract(r'(\d+)x(\d+)').astype(int)
df.drop(columns=["ScreenResolution"], inplace=True)
df.head()
```

```python
df[['CPU', 'Frequency']] = df['Cpu'].str.rsplit(' ', n=1, expand= True)
```

```python
df[['Storage', 'StorageCapacity']] =df['Memory'].str.rsplit(' ', n=1, expand= True)
```

```python
df
```

```python
df['Frequency']=df['Frequency'].str.replace('GHz', ''). astype(float)
df
```

```python
valid_companies = ["Apple", "Dell", "HP", "Lenovo", "Asus", "Acer", "MSI", "Toshiba"]
df["Company"] = df["Company"].where(df["Company"].isin(valid_companies), "Other")
df["Company"].value_counts()
```

```python
import seaborn as sns
import matplotlib.pyplot as plt

plt.figure(figsize=(6, 4))
sns.boxplot(x=df['Inches'])
plt.title("Boxplot")
plt.show()
```

```python
df['Inches'].describe()
```

```python
Q1 = df['Inches'].quantile(0.25)
Q3 = df['Inches'].quantile(0.75)
IQR = Q3 - Q1

lower_bound = Q1 - 1.5 * IQR
upper_bound = Q3 + 1.5 * IQR

mean_value = df['Inches'].mean()

df['Inches'] = np.where((df['Inches'] < lower_bound) | (df['Inches'] > upper_bound),mean_value,df['Inches'])
print(df['Inches'])
```

```python
import seaborn as sns
import matplotlib.pyplot as plt

plt.figure(figsize=(6, 4))
sns.boxplot(x=df['Inches'])
plt.title("Boxplot")
plt.show()
```
